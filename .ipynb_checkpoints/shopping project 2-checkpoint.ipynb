# Univariate analysis (one variable)

```python
df.describe()
```

Based on the above table, we can say that based on the 200 customers, the average income is roughly 60.6k a year and that the median income is 61k. This distribution may be a bit of a normal distribution based on the values being similar.

```python
pnd.distplot(df['Annual Income (k$)'])
```

```python
str(df.columns); # create a list of table columns
```

```python
columns = ['Age', 'Annual Income (k$)','Spending Score (1-100)']
```

```python
for i in columns:
    plt.figure() # create a new figure for each visual instead of stacking them
    pnd.distplot(df[i]) # create a new visual histogram
```

```python
pnd.kdeplot(shade=True,x=df['Annual Income (k$)'],hue=df['Gender']);
# We can kplot the Annual income, but we can shade based on Gender
```

Lets create the same type of graphs as before as KDE plots. KDE Plot is a way to show how many things you have of different sizes through the curves vs boxes in a histogram.

```python
for i in columns:
    plt.figure()
    pnd.kdeplot(shade=True,x=df[i],hue=df['Gender']);
```

We can that a majority of customers are females mid 20s to mid 40s. Annual income is also higher for this range as well as spending score.

Below are some box plots of the data.

```python
for i in columns:
    plt.figure()
    pnd.boxplot(data=df,x='Gender',y=df[i])
```

We can see that the box plot for age VS gender, the median age for a male shopper is higher and there is a wider range of male shoppers age vs their female counter parts. In the second chart, we can see that there is a male outlier which caused a tail increase in the KDE plot for annual income.
